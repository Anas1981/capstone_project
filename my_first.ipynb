## Columns label in details 
* MATCH_DATE = The date when the match was played.
* LEAGUE_NAME =The name of the league in which the match was played.
* SEASON = The season(year) during which the match was played.
* LEAGUE = The abbreviation of the league name.
* FINISHED = A boolean indicating whether the match was finished or not.
* LOCATION = The location where the match was played.
* VIEWER = The number of viewers who watched the match.
* MATCHDAY_NR = The matchday number within the season as int
* MATCHDAY =The matchday number within the season as string.
* HOME_TEAM_ID = The unique identifier of the home team.
* HOME_TEAM_NAME =The name of the home team.
* HOME_ICON = The icon of the home team as url.
* AWAY_TEAM_ID =The unique identifier of the away team.
* AWAY_TEAM_NAME = The name of the away team.
* AWAY_TEAM =The abbreviation of the away team name.
* AWAY_ICON = The icon of the away team as url.
* GOALS_HOME = The number of goals scored by the home team.
* GOALS_AWAY = The number of goals scored by the away team.
* DRAW = A boolean (0 or 1) indicating whether the match ended in a draw or not.
* WIN_HOME =A boolean (0 or 1) indicating whether the home team won the match or not.
* WIN_AWAY =A boolean (0 or 1) indicating whether the away team won the match or not.
